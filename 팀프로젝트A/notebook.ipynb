````python
df.head()
````

````python
df_group = df.groupby('월').agg({'거래ID' : 'size', '고객ID' : 'nunique', '총금액' : 'sum'})
df_group.columns = ['거래 수','이용고객 수', '총금액']
df_group['거래 당 구매금액'] = round(df_group['총금액'] / df_group['거래 수'], 3)
df_group['고객 당 구매금액'] = round(df_group['총금액'] / df_group['이용고객 수'], 3)

styled_all_columns = df_group.style.background_gradient(cmap='coolwarm', subset=['거래 수', '거래 당 구매금액'])
styled_all_columns
````

> 3개의 그래프를 하나로 나타내기

````python
monthly_summary = df.groupby(pd.Grouper(key='거래날짜', freq='MS')).agg({'수량': 'sum', '총금액': 'sum', '거래ID' : 'size'}).rename(columns = {'거래ID' : '거래건수'})
monthly_summary['수량_상대크기'] = monthly_summary['수량'].rank(ascending=False)
monthly_summary
````

````python
monthly_summary = df.groupby('월').agg({'수량': 'sum', '총금액': 'sum', '거래ID' : 'size'}).rename(columns = {'거래ID' : '거래건수'})
monthly_summary['수량_상대크기'] = monthly_summary['수량'].rank(ascending=False)
# 그래프 그리기
plt.figure(facecolor='white', figsize=(10,10))
fig, ax1 = plt.subplots(facecolor='white')

# 막대 그래프 (거래건수)
ax1.bar(monthly_summary.index,monthly_summary['거래건수'], color='b', alpha=0.5, label='거래건수', width=0.5)
ax1.set_xlabel('월')
ax1.set_ylabel('거래건수', color='b')
ax1.tick_params('y', colors='b')
ax1.set_ylim(3000, 9000)

# 라인 그래프 (매출액)
ax2 = ax1.twinx()
ax2.plot(monthly_summary.index, monthly_summary['총금액'], color='r', marker='o', label='매출액')
ax2.set_ylabel('매출액', color='r')
ax2.tick_params('y', colors='r')
ax2.set_ylim(0, 550000)

# 라인 그래프 (판매량)
ax2.scatter(monthly_summary.index, monthly_summary['총금액'], color='g', s= [i*30 for i in monthly_summary['수량_상대크기']], label='판매량')

# 범례 표시
ax1.set_facecolor('white')
ax2.set_facecolor('white')
# ax1.legend(loc='upper left')
ax2.legend(loc='upper left')

plt.title('월별 거래건수, 매출액, 판매량')
plt.show()
````

### 월별 온라인/오프라인 마케팅 비용 및 거래건수, 총거래액 동향

### RFM 분석
--------------------
점수별로 고객의 등급을 부여

▶50점 이상 A\
▶40 ~ 50점 B\
▶20 ~ 40점 C\
▶10 ~ 20점 D\
▶0 ~ 10점 F

지역은 시카고, 워싱턴

````python
# Recency 계산
df['recency'] = (df['거래날짜'].max() - df['거래날짜']).dt.days

# Frequency 계산
frequency_df = df.groupby('고객ID').agg({'거래ID': 'nunique'}).rename(columns={'거래ID': 'frequency'})

# Monetary 계산
monetary_df = df.groupby('고객ID').agg({'총금액': 'sum'}).rename(columns={'총금액': 'monetary'})

# RFM 데이터프레임 합치기
rfm_df = frequency_df.join(monetary_df)

# 각 고객별 가장 최근 구매일
recency_df = df.groupby('고객ID')['거래날짜'].max().reset_index()
recency_df['recency'] = (df['거래날짜'].max() - recency_df['거래날짜']).dt.days

# 최종 RFM 데이터프레임
rfm_df = rfm_df.join(recency_df.set_index('고객ID'))

# 필요한 열만 선택
rfm_df = rfm_df[['recency', 'frequency', 'monetary']]

# 결과 확인
rfm_df.head()
````

recency : 최근성

frequency : 구매빈도

monetary : 총금액

### score별 등급 기준 선정

````python
# ▶ Min max scale = 최대값을 1, 최소값을 0으로 표준화하는 기법
from sklearn.preprocessing import minmax_scale


# ▶ 최근성은 숫자가 작을수록, 즉 최근 구매일이 얼마 지나지 않은 고객이 더 점수가 높음
rfm_df['recency'] = minmax_scale(rfm_df['recency'], axis=0, copy=True)
rfm_df['recency'] = 1-rfm_df['recency']

rfm_df['frequency'] = minmax_scale(rfm_df['frequency'], axis=0, copy=True)

rfm_df['monetary'] = minmax_scale(rfm_df['monetary'], axis=0, copy=True)

rfm_df['score']=rfm_df['recency']+rfm_df['frequency']+rfm_df['monetary']

# ▶ Score
rfm_df['score']=minmax_scale(rfm_df['score'], axis=0, copy=True)*100
rfm_df['score']=round(rfm_df['score'],0)

rfm_df.head()
````

````python
# ▶ 점수별로 고객의 등급을 부여
'''
50점 이상 A
40~50점 B
20~40점 C
10~20점 D
0~10점 F
'''
rfm_df['grade'] = rfm_df['score'].apply(lambda x : 'A' if x>=50 else
                                  ('B' if x>=40 else
                                   ('C' if x>=20 else
                                    'D' if x>=10 else 'F')))
rfm_df.head()
````

2. Washington DC의 구매력을 높이기 위해 해당 지역의 A~C등급의 고객들이 선호하는 상위 제품에 대한 추가 할인쿠폰을 제공하여 고객들의 충성도를 강화시킨다. 이를 통해 충성 고객의 이탈 방지 효과와 기존 고객들의 구매력을 높일 수 있다.

3. Washington DC 지역은 이용 고객 수가 적은 반면, 높은 등급 고객의 비율이 높은  지역이다. 신규 고객을 더 늘린다면 다른 지역보다 수익률이 높을 것으로 예상되므로, 오프라인 마케팅을 이 지역에 집중하는 것을 제안한다.

4. 마케팅 비용이 증가할수록 총 거래액이 같이 증가하는 경향을 보이는 것을 알 수 있었다. 12월이 온라인/오프라인 마케팅 비용이 가장 많았으며, 동시에 총 거래액도 가장 많았으므로, 4분기에 프로모션을 집중한다면 효과인 매출 상승을 기대할 수 있을 것이다.
