### RFM Î∂ÑÏÑù
--------------------
Ï†êÏàòÎ≥ÑÎ°ú Í≥†Í∞ùÏùò Îì±Í∏âÏùÑ Î∂ÄÏó¨

‚ñ∂50Ï†ê Ïù¥ÏÉÅ A\
‚ñ∂40 ~ 50Ï†ê B\
‚ñ∂20 ~ 40Ï†ê C\
‚ñ∂10 ~ 20Ï†ê D\
‚ñ∂0 ~ 10Ï†ê F

ÏßÄÏó≠ÏùÄ ÏãúÏπ¥Í≥†, ÏõåÏã±ÌÑ¥

````python
# Recency Í≥ÑÏÇ∞
df['recency'] = (df['Í±∞ÎûòÎÇ†Ïßú'].max() - df['Í±∞ÎûòÎÇ†Ïßú']).dt.days

# Frequency Í≥ÑÏÇ∞
frequency_df = df.groupby('Í≥†Í∞ùID').agg({'Í±∞ÎûòID': 'nunique'}).rename(columns={'Í±∞ÎûòID': 'frequency'})

# Monetary Í≥ÑÏÇ∞
monetary_df = df.groupby('Í≥†Í∞ùID').agg({'Ï¥ùÍ∏àÏï°': 'sum'}).rename(columns={'Ï¥ùÍ∏àÏï°': 'monetary'})

# RFM Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ìï©ÏπòÍ∏∞
rfm_df = frequency_df.join(monetary_df)

# Í∞Å Í≥†Í∞ùÎ≥Ñ Í∞ÄÏû• ÏµúÍ∑º Íµ¨Îß§Ïùº
recency_df = df.groupby('Í≥†Í∞ùID')['Í±∞ÎûòÎÇ†Ïßú'].max().reset_index()
recency_df['recency'] = (df['Í±∞ÎûòÎÇ†Ïßú'].max() - recency_df['Í±∞ÎûòÎÇ†Ïßú']).dt.days

# ÏµúÏ¢Ö RFM Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ
rfm_df = rfm_df.join(recency_df.set_index('Í≥†Í∞ùID'))

# ÌïÑÏöîÌïú Ïó¥Îßå ÏÑ†ÌÉù
rfm_df = rfm_df[['recency', 'frequency', 'monetary']]

# Í≤∞Í≥º ÌôïÏù∏
rfm_df.head()
````

### scoreÎ≥Ñ Îì±Í∏â Í∏∞Ï§Ä ÏÑ†Ï†ï

````python
# ‚ñ∂ Min max scale = ÏµúÎåÄÍ∞íÏùÑ 1, ÏµúÏÜåÍ∞íÏùÑ 0ÏúºÎ°ú ÌëúÏ§ÄÌôîÌïòÎäî Í∏∞Î≤ï
from sklearn.preprocessing import minmax_scale


# ‚ñ∂ ÏµúÍ∑ºÏÑ±ÏùÄ Ïà´ÏûêÍ∞Ä ÏûëÏùÑÏàòÎ°ù, Ï¶â ÏµúÍ∑º Íµ¨Îß§ÏùºÏù¥ ÏñºÎßà ÏßÄÎÇòÏßÄ ÏïäÏùÄ Í≥†Í∞ùÏù¥ Îçî Ï†êÏàòÍ∞Ä ÎÜíÏùå
rfm_df['recency'] = minmax_scale(rfm_df['recency'], axis=0, copy=True)
rfm_df['recency'] = 1-rfm_df['recency']

rfm_df['frequency'] = minmax_scale(rfm_df['frequency'], axis=0, copy=True)

rfm_df['monetary'] = minmax_scale(rfm_df['monetary'], axis=0, copy=True)

rfm_df['score']=rfm_df['recency']+rfm_df['frequency']+rfm_df['monetary']

# ‚ñ∂ Score
rfm_df['score']=minmax_scale(rfm_df['score'], axis=0, copy=True)*100
rfm_df['score']=round(rfm_df['score'],0)

rfm_df.head()
````
