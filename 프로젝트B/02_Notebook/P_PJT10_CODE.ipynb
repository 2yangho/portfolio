````python
#ìµœê·¼ 5ê°œë…„ ì™„ê²° ì›¹íˆ°, ì—°ì¬ì¤‘ ì›¹íˆ°ë§Œ ì¡°ì‚¬
df1_pivot = df1[(df1.newdate.dt.year >= 2019) & (df1.completed == True)].pivot_table(index=df1.newdate.dt.year, columns='genre', aggfunc='size')
df1_pivot.fillna(0, inplace=True)
df1_pivot.loc['ì—°ì¬ì¤‘'] = df1[df1.completed == False].pivot_table(index=df1.newdate.dt.year, columns='genre', aggfunc='size').values[0]
df1_pivot.loc['í•©ê³„'] = df1_pivot.sum()
df1_pivot['ê³„'] = df1_pivot.sum(axis=1)
df1_pivot
````

````python
df1_pivot_ratio = df1_pivot.copy()
for i in range(len(df1_pivot_ratio.columns)-1):
    df1_pivot_ratio.iloc[:,i] = df1_pivot_ratio.iloc[:,i] / df1_pivot_ratio['ê³„']
df1_pivot_ratio.drop('ê³„', axis=1, inplace=True)
df1_pivot_ratio
````

````python
import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import seaborn as sns

#í•œê¸€í°íŠ¸ ì ìš©
font_path = 'C:\\Windows\\Fonts\\gulim.ttc'
font = fm.FontProperties(fname=font_path).get_name()
mpl.rc('font', family=font)

df1_pivot_ratio.index = df1_pivot_ratio.index.astype(str)   # indexì— ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬í•˜ê¸° ì „ì— ëª¨ë“  indexì˜ ë°ì´í„°íƒ€ì… ë§ì¶°ì£¼ê¸°

ax = df1_pivot_ratio.sort_index(ascending=False).plot(kind='barh', stacked=True, title="ì—°ë„ë³„ ì¸ê¸°ì¥ë¥´ë¹„ìœ¨", rot=0);
for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate("%.1f"%(width*100), xy=(left+width/2, bottom+height/2), ha='center', va='center', color='r');

plt.box(False)
plt.gcf().set_size_inches(10, 5)
plt.show()
````
