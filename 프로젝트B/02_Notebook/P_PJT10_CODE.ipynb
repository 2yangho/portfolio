### 1. 연도별 인기 장르 조사
---

````python
#최근 5개년 완결 웹툰, 연재중 웹툰만 조사
df1_pivot = df1[(df1.newdate.dt.year >= 2019) & (df1.completed == True)].pivot_table(index=df1.newdate.dt.year, columns='genre', aggfunc='size')
df1_pivot.fillna(0, inplace=True)
df1_pivot.loc['연재중'] = df1[df1.completed == False].pivot_table(index=df1.newdate.dt.year, columns='genre', aggfunc='size').values[0]
df1_pivot.loc['합계'] = df1_pivot.sum()
df1_pivot['계'] = df1_pivot.sum(axis=1)
df1_pivot
````

````python
df1_pivot_ratio = df1_pivot.copy()
for i in range(len(df1_pivot_ratio.columns)-1):
    df1_pivot_ratio.iloc[:,i] = df1_pivot_ratio.iloc[:,i] / df1_pivot_ratio['계']
df1_pivot_ratio.drop('계', axis=1, inplace=True)
df1_pivot_ratio
````

````python
import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import seaborn as sns

#한글폰트 적용
font_path = 'C:\\Windows\\Fonts\\gulim.ttc'
font = fm.FontProperties(fname=font_path).get_name()
mpl.rc('font', family=font)

df1_pivot_ratio.index = df1_pivot_ratio.index.astype(str)   # index에 내림차순 정렬하기 전에 모든 index의 데이터타입 맞춰주기

ax = df1_pivot_ratio.sort_index(ascending=False).plot(kind='barh', stacked=True, title="연도별 인기장르비율", rot=0);
for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate("%.1f"%(width*100), xy=(left+width/2, bottom+height/2), ha='center', va='center', color='r');

plt.box(False)
plt.gcf().set_size_inches(10, 5)
plt.show()
````

로맨스 장르가 가장 인기가 많으며, 그다음으로 드라마/판타지 장르가 꾸준히 인기가 많다. 그 다음 인기 장르는 액션, 스릴러이고, 비인기장르는 감성, 개그, 무협/사극, 스포츠, 일상 장르이다. 

연도에 따라 뚜렷한 증가 비율을 보이는 장르는 로맨스이고, 비율 감소를 보이는 장르는 개그이다. 

추가로, 로맨스/ 드라마 장르가 인기장르인것으로 보았을 때 네이버웹툰은 여성이용자가 많은 듯 하다.

### 3. 작가별 작품수 및 평점평균
---

````python
df1.head()
````

몇몇 작품은 다수의 작가가 작업한 것인데, 해당 작품이 기록된 필드는 기여한 작가 수만큼 복제한 뒤, 각 복제 필드마다 작가이름을 분할하여 넣는다. 

````python
df1_author = df1.copy()
#authlist = []

for id, rows in enumerate(df1_author.author):
    if "/" in rows:
        for auth in rows.split(' / ')[1:]:
            # print(auth)
            # authlist.append(auth)
            new_row_index = str(id)+auth
            #print(new_row_index)
            df1_author.loc[new_row_index] = df1.loc[id]
            df1_author.loc[new_row_index, 'author'] = auth
            #print(df1_author.loc[new_row_index].author)
            

        df1_author.loc[id, 'author'] = rows.split(' / ')[0]


df1_group = df1_author.groupby('author').size().sort_values(ascending=False).to_frame()
df1_group.rename(columns={0:'size'}, inplace=True)
df1_group.head()
````

아쉬운 점이 있다면, 두 웹툰 데이터셋에 작품의 흥행도를 확인할 수 있는 지표가 별점밖에 없다는 사실이다. 만약에 작품별 조회수, 댓글수 데이터가 있었다면 EDA 과정을 더 깊게 진행할 수 있었을 것이다. 해당 프로젝트에서는 별점밖에 없다보니 프로젝트의 목표를 충족시키기에는 다소 부족한 점이 있었다.
